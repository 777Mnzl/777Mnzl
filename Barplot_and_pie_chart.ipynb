```python
party = [ 'RPP', 'RSP','Congress','CPN-UML','Maoist']
seats = [21,23,22,40,20]
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#plotly
px.bar(x = party, y = seats)
```

```python
#seaborn
sns.barplot(x = party, y = seats) #bar chart seaborn bata ramro huncha
```

```python
#matplotlib
plt.bar(x = party, height = seats)
plt.title('Party and their seat counts')
plt.ylabel('No. of seats')
```

```python
px.pie(names = party, values = seats) #pie chart haru plotly bata ramro huncha
```

```python
plt.pie(x = seats, labels = party)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
 from sklearn.datasets import load_iris
 X, y = load_iris(return_X_y = True, as_frame = True)
X.head()
```

```python
plt.boxplot(X['petal length (cm)'] )
plt.xlabel('petal')
plt.ylabel('length')
plt.show()
```

```python
plt.boxplot(X['sepal width (cm)'] )
plt.xlabel('sepal')
plt.ylabel('width (cm)')
plt.show()
```

```python
sns.boxplot(y = X['petal width (cm)'], color = 'green', linewidth = 5.6)
```

```python
for col in X.columns:
  plt.figure(figsize = (3,3)) #figure ko size increase or decrease garna milcha
  plt.title(f'{col}') #title dekhaucha
  sns.boxplot(X[col], color = 'red')

  plt.show()
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
X, y = load_iris(return_X_y=True, as_frame = True)
X.head()
```

```python
plt.scatter(X['sepal width (cm)'], y = X['petal length (cm)'])
```

```python
sns.scatterplot(x = X['sepal length (cm)'], y=X['petal width (cm)'])
plt.title('relation between sepal length and petal width')
plt.xlabel('sepal length (cm)')
plt.ylabel('petal width(cm)')
```

```python
fig = px.scatter(x = X['sepal length (cm)'], y=X['petal width (cm)'])
fig.update_layout(
    title = 'Scatterplot: relatio between sepal length and petal width',
    xaxis_title = 'sepal length',
    yaxis_title = 'petal width',
)
```
