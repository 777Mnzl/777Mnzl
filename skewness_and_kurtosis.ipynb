```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.normal(loc=5, scale=1, size=2000)
```

```python
plt.figure(figsize =(5,3))
sns.histplot(data, kde =True)
```

```python
from scipy.stats import skew
skew(data)
```

# Kurtosis
1. Mesokurtic - normal distribution

```python
np.random.seed(40)
data = np.random.standard_normal(size=3000)
plt.figure(figsize = (5,3))
sns.histplot(data, kde = True)
```

```python
from scipy.stats import kurtosis
kurtosis(data, fisher = False) # pearson method
```

```python
kurtosis(data, fisher =True) # Fisher method
```

# Platykurtic distributions hace kurtosis < 3
* uniform distribution

```python
np.random.seed(40)
data = np.random.uniform(low = 2, high = 6, size = 30000)
plt.figure(figsize = (5,3))
sns.histplot(data, kde = True)
```

```python
kurtosis(data, fisher = False) # pearson method less than 3
```

```python
kurtosis(data, fisher = True)
```

# A leptokurtic distribution is fat-tailed, meaning , meaning that there are a lot of outliers. Kurtosis > 3

```python
np.random.seed(40)
data = np.random.laplace(size = 30000)
plt.figure(figsize = (5,3))
sns.histplot(data, kde = True)
```

```python
plt.figure(figsize = (5,3))
sns.kdeplot(data)
```

```python
from scipy.stats import kurtosis
kurtosis(data, fisher = False)# pearson method greater than 3 -- has outliers
```
